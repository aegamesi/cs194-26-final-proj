```python
%matplotlib
import os
import re
import io
import time
import functools
import json

import numpy as np
import skimage.io as skio
import skimage as sk
import IPython
import PIL.Image
import IPython.display
import scipy.ndimage
import skimage.transform
import matplotlib.pyplot as plt
import cv2

def imread(filename):
    im = skio.imread(filename)
    im = sk.img_as_float(im)
    return im

def imsave(filename, im):
    im = np.clip(im, -1, 1)
    skio.imsave(filename, im)

def showimage(im, clip=True):
    if clip:
        im = np.clip(im, 0, 1)
    im = sk.img_as_ubyte(im)
    f = io.BytesIO()
    PIL.Image.fromarray(im).save(f, 'jpeg')
    IPython.display.display(IPython.display.Image(data=f.getvalue()))
```

```python
def correspondences(im1, im2):
    pts1 = []
    pts2 = []
    while True:
        plt.imshow(im1)
        plt.draw()
        plt.axis('image')
        p = plt.ginput(n=1, timeout=0, show_clicks=True)
        if not p:
            break
        pts1.extend(p)
            
        plt.imshow(im2)
        plt.draw()
        plt.axis('image')
        p = plt.ginput(n=1, timeout=0, show_clicks=True)
        if not p:
            break
        pts2.extend(p)
        
        print("got {} points".format(len(pts1)))
    print("done")
    return pts1, pts2
```

```python
def combine(a, b):
    return np.multiply(a, (1 - np.ma.masked_where(b != 0, b).mask)) + b
```
